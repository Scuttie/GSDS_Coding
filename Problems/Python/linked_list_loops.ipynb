# 4. Linked List Loops (7 pts.)
Write a function that determines if there is a loop in a singly linked list. A loop is present when the last node of the linked list connects to a previous node. Values in the linked list are not unique.

- **Requirements**
  * Function `hasLoop` returns two values in the following order
    - Length of the list (int)
    - A boolean indicating if there is a loop: True if there is a loop and False
  * You cannot modify the linked list.

- **Example**

The following figure shows a linked list with a loop where the last node connects back to the 3rd node.

image.png

```python
# Do not modify!
class LinkedNode:
  def __init__(self, value):
    self.value = value
    self.next = None
```

```python
def hasLoop(head):
  #Write your code here
    #지금까지 방문한 곳을 저장하는 리스트 
    visited_list = []

    #head부터 시작해서 방문 시작
    explorer = head

    #explorer visited_list에 넣기
    visited_list.append(explorer)

    #return 될때까지 반복
    while True:
        try:
            #explorer의 다음 노드로 향하기
            explorer = explorer.next
            #만약에 explorer가 visited_list에 있으면 (loop가 발생하면)
            if explorer in visited_list:
                #loop가 있음을 알리고 visited_list의 길이를 반환
                return len(visited_list), True
            #loop가 발생하지 않으면
            else:
                #visited_list에 넣고 다음 루프로 진행
                visited_list.append(explorer)
        except:
            #다음 노드로 향하지 못하는 경우 (linked list의 끝까지 도달)
            #visited_list의 length - 1 반환 (마지막 노드는 NULL)하고 False 반환
            return len(visited_list) - 1, False
            break
```

```python
head = LinkedNode(1)
curr = head
vals = [2,3,4,5]
for i in vals:
  new = LinkedNode(i)
  curr.next = new
  curr = new
print(hasLoop(head)) #5, False
curr.next = head.next
print(hasLoop(head)) #5, True
```
